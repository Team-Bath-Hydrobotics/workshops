# Task
Given our Architecture displayed in existingArchitecture.png, we'd like you to answer some questions, answer whichever questions you want to. Following this feel free to freestyle and come up with any of your own modifications. There are no right or wrong answers, we just want to hear your thought process.

You might like to use diagrams, where appropriate, a good starting place could be Mermaid [https://mermaid.js.org/] or Draw.io [https://app.diagrams.net/]

Feel free to present your solutions as a presentation or document, or any other way you see fit.

# Questions

## 1. Bottleneck Analysis
What potential bottlenecks do you see in our design? Consider:
- Network bandwidth and latency
- Processing power on the ROV vs. surface station
- Data storage and transmission

---

## 2. Multi-ROV Scalability
If you had to handle multiple ROVs at once, what might need to change about our design?
- How would you manage multiple video streams?
- What changes to the UI/UX would be needed?
- What changes to data storage might be needed ?
---

## 3. Autonomous Operation
If our ROV was automated, how might you change the design?
- What new components would you add?
- How would autonomous decision-making integrate with the current architecture?
- What role would the surface station play?
- How would you handle fail-safes and emergency override?

---

## 4. Fault Tolerance & Reliability
What happens when things go wrong?
- How would you handle the broker going down ?
- How would you handle losing connection to the ROV ?
---

## 5. Data Management Strategy
How would you handle the massive amounts of data an ROV generates?
- Real-time vs. recorded data - what gets stored where?
- How long should mission data be retained?
- What metadata is important to capture?

---

## 6. Security & Access Control
Who should be able to control the ROV and access its data?
- How would you implement authentication?
- What permission levels make sense?
- How do you secure the communication channel?
- What about physical security of the equipment?
